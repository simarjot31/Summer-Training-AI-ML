```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
import numpy as np
total_samples = 61

# Initialize arrays with flexible shape for images
data = np.zeros((total_samples, 512*512))
label = np.zeros((total_samples))
images = [] # Store images in a list initially

label[40:] = 1

print(data.shape)
print(label.shape)

for i in range(1,total_samples+1,1):
  path = "/content/drive/MyDrive/Colab Notebooks/Day 7/new_train-20240708T092551Z-001/new_train/d(%d).png"%(i)
  im = mimg.imread(path)

  temp_im = im[:,:,0]
  feat = temp_im.reshape(1,-1)
  data[i-1,:] = feat # Assign the features to data array
  images.append(temp_im) # Append each image to the list

# Convert the list of images to a NumPy array after the loop
images = np.array(images)
print(images.shape)
print(label)
```

```python
from sklearn import svm, metrics, model_selection

x = data
y = label
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=0.3,random_state=4)

tumor_model = svm.SVC(kernel='linear', gamma="scale", C=2)
tumor_model.fit(xtrain,ytrain)

ypred = tumor_model.predict(xtest)

print(metrics.classification_report(ypred,ytest))
print(metrics.accuracy_score(ypred,ytest))
```
