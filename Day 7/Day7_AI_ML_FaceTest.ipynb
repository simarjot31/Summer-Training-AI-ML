```python
import matplotlib.image as mimg
import matplotlib.pyplot as plt
import numpy as np
total_samples = 61
data = np.zeros((total_samples,512*512))
label = np.zeros((total_samples))
images = np.zeros((total_samples,512,512))

label[40:] = 1

print(data.shape)
print(label.shape)
print(images.shape)

for i in range(1,total_samples+1,1):
  path = "/gdrive/MyDrive/AI_Summer_Training/data_files/brain_tumor/d(%d).png"%(i)
  im = mimg.imread(path)

  temp_im = im[:,:,0]

  feat = temp_im.reshape(1,-1)

  data[i-1,:] = feat
  images[i-1,:,:] = temp_im

print(label)
```

```python
from sklearn import svm, metrics, model_selection

x = data
y = label
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(x,y,test_size=0.3,random_state=4)

tumor_model = svm.SVC(kernel='linear', gamma="scale", C=2)
tumor_model.fit(xtrain,ytrain)

ypred = tumor_model.predict(xtest)

print(metrics.classification_report(ypred,ytest))
print(metrics.accuracy_score(ypred,ytest))
```
