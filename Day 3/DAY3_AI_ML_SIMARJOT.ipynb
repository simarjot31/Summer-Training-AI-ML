```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/ign.csv'
data = pd.read_csv(path)
print(data.shape)
data.head()
```

```python
# @title Number of Games Released on Each Platform

import matplotlib.pyplot as plt

# Count the number of games on each platform
platform_counts = data['platform'].value_counts()

# Create a bar plot
plt.bar(platform_counts.index, platform_counts.values)
plt.xlabel('Platform')
plt.ylabel('Number of Games')
plt.title('Number of Games Released on Each Platform')
_ = plt.xticks(rotation=90)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

path = '/content/Pokemon.csv'
data = pd.read_csv(path)
print(data.shape)
data.head()
```

```python
# remove the hash column

newData = data.drop(['#'],axis=1, inplace=False) #it will replace the hash column and update the data variable itself
# newData = data.drop(['#'],axis=1, inplace=True)

newData.columns
```

```python
col = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
plt.figure(1,figsize=(15,10))
sns.boxplot(data=newData[col])
```

```python
#Distribution plot

plt.figure(1,figsize=(10,8))

plt.subplot(2,2,1)
sns.histplot(data=newData, x='Attack')

plt.subplot(2,2,2)
sns.histplot(data=newData, x='HP')

plt.subplot(2,2,3)
sns.histplot(data=newData, x='Speed')

plt.subplot(2,2,4)
sns.histplot(data=newData, x='Defense')
```

```python
#Display the count plot of type 1 pokemon

plt.figure(1,figsize=(17,6))
ax = sns.countplot(data=newData, x='Type 1')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
```

```python
#Display the count plot of type 2 pokemon

plt.figure(1,figsize=(17,6))
ax = sns.countplot(data=newData, x='Type 2')
for i in (ax.containers):
  ax.bar_label(i)
#plt.xticks(rotation=90)
```

```python
#Display the number of pokemon in each generation
#Show it in a virtual form

plt.figure(1,figsize=(17,6))
ax = sns.countplot(data=newData, x='Generation')
for i in (ax.containers):
  ax.bar_label(i)

plt.xticks(rotation=90)
```

```python
#Compare the type 1 pokemon with repect to attack capability

plt.figure(1,figsize=(17,6))
ax = sns.barplot(data=newData, x='Type 1', y='Attack')
for i in (ax.containers):
  ax.bar_label(i)
```

```python
#Compare the type 2 pokemon based upon their speed capabilty

plt.figure(1,figsize=(17,6))
ax = sns.barplot(data=newData, x='Type 2', y='Speed')
for i in (ax.containers):
  ax.bar_label(i)
```

```python
#Compare the type 2 pokemon generation wise based upon their speed capabilty

plt.figure(1,figsize=(17,6))
ax = sns.barplot(data=newData, x='Type 2', y='Speed' , hue='Generation', errorbar=None)
for i in (ax.containers):
  ax.bar_label(i)

plt.xticks(rotation=90)
```

```python
#Compare out the legendary and non-legendary pokemon
#then perform the analysis on legendary and non-legendary pokemons separately

#compare the attack, defence and speed with respect to their types
#compare their total strength with respect to generation

ind1= data['Legendary'] == True  #For Legendary
ind2 = data['Legendary'] == False   #For non-legendary

legend_true=data[ind1]  #This will hold the data for legendary data
legend_false=data[ind2]  ##This will hold the data for non-legendary data

print("total number of legendary pokemon are: ",legend_true.shape[0])
print("total number of non-legendary pokemon are: ",legend_false.shape[0])
```

```python
plt.figure(1,figsize=(15,6))
plt.subplot(1,2,1)
ax = sns.barplot(data=legend_true, x='Type 1', y='Speed')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Legendary Pokemon Speed Analysis')

plt.subplot(1,2,2)
sns.barplot(data=legend_false, x='Type 1', y='Speed')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Non-Legendary Pokemon Speed Analysis')
```

```python
plt.figure(1,figsize=(15,6))
plt.subplot(1,2,1)
ax = sns.barplot(data=legend_true, x='Type 1', y='Defense')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Legendary Pokemon Defense Analysis')

plt.subplot(1,2,2)
sns.barplot(data=legend_false, x='Type 1', y='Defense')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Non-Legendary Pokemon Defense Analysis')
```

```python
plt.figure(1,figsize=(15,6))
plt.subplot(1,2,1)
ax = sns.barplot(data=legend_true, x='Type 1', y='Attack')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Legendary Pokemon Attack Analysis')

plt.subplot(1,2,2)
sns.barplot(data=legend_false, x='Type 1', y='Attack')
for i in (ax.containers):
  ax.bar_label(i)
plt.xticks(rotation=90)
plt.title('Non-Legendary Pokemon Attack Analysis')
```
