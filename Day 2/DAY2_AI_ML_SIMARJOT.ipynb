```python
# write a program to create an autothreshold algorithm
# 1). create a 9x9 matrix within a raneg of 0-100 integer number
# 2). find the average of each row and save it in a B list (1x9)
#step 2: [B0, B, B2, 83, 84, 85, 86, 87, 88]

# 3). NOW, COMPARE the elements of each row with its respective #mean. if mean value is greater than the element, then leave the
# element as it is else make the element 0 create a new matrix C # to hold the updated result

#4). count the number of zeros and non zeros elements, which should
# be equal to the number of elements in a matrix


#step 1: creating an array of 9x9
np.random.seed(5)
A= np.random.randint(0,100, size=(9,9))
print("Input Matrix")
print(A.shape)
print(A)
print()

#step 2
B = []     #empty list to save the mean
for i in range(0,A.shape[0],1):
 temp= A[i,:].mean()
 B.append(temp)  #save the mean on the respective positioon of the B list
print("Output Matrix")
print(B)

print()

#step 3 and step 4
C = np.zeros((A.shape[0],A.shape[1]))

ZerosCount=0
Non_Zero_Count=0

for i in range(A.shape[0]):   #row iteration
  for j in range(A.shape[1]):   #column iteration
    if (A[i,j])>B[i]:
      C[i,j]=0
      ZerosCount=ZerosCount+1
    else:
      C[i,j]=A[i,j]
      Non_Zero_Count=Non_Zero_Count+1

if (Non_Zero_Count+ZerosCount == A.size):
  print("Updated Matrix is verifed")
else:
  print("Updated Matrix is not verifed")

print()

print("Updated Matrix")
print(C)
print()
print("No. of Zeros Count: ", ZerosCount)
print("Non Zero Count:", Non_Zero_Count)

```

```python
#Slicing sub-matrix from the matrix

import numpy as np
np.random.seed(5)
sz = 8
A= np.random.randint(10,20, size=(sz,sz))
print("Input Matrix: ",A.shape)
print(A)

print("\n")

Result=np.zeros((int(A.shape[0]/2),int(A.shape[1]/2)))
indi=-1
indj=-1

for i in range(0,A.shape[0],2):
    indi=indi+1
    indj=-1
    for j in range(0,A.shape[1],2):
        indj=indj+1
        s=A[i:i+2, j:j+2]
        print()
        print(s)
        temp = s
        max_value = temp.max()
        print("Max value: ", max_value)
        Result[indi,indj] = max_value
        print()
print("Final Output:")
print(Result)
```
