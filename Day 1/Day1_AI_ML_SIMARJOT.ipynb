```python
#write a progam to perform arithmetic operation(+,-,/,*) by taking ipt form the user

flag=0
while(flag==0):
  a= int(input("number 1:"))
  b= int(input("number 2:"))

  print("Enter operator(+,-,*,/)")
  operator=input('operator:')


  if operator=='+':
     print(a+b)
  elif operator=='-':
     print(a-b)
  elif operator=='*':
     print(a*b)
  elif operator=='/':
     print(a/b)
  else:
     print("Wrong Input")

  check=input("Do you want to continue(y/n):")
  if(check=='n'):
    flag=1 #change the flag status
    print("Exit")
  elif(check=='y'):
    flag=0
    print("Program Continue...")
```

```python
def input_op():
  A = int(input("number 1:"))
  B = int(input("number 2:"))

  flag=0
  while(flag==0):
    print("Enter operator(+,-,*,/)")
    op=input('operator:')
    if (op=='*' or op=='/' or op=='+' or op=='-'):
      flag=1
      print("right operator selected")
    else:
      flag=0
      print("Wrong Input")
  return([A,B,op])
```

```python
def execution_func(input_list):
  a=input_list[0]
  b=input_list[1]
  op=input_list[2]
  if(op=="+"):
    c=a+b
    print("Result:",c)
  elif(op=="-"):
    c=a-b
    print("Result:",c)
  elif(op=="*"):
    c=a*b
    print("Result:",c)
  elif(op=="/"):
    c=a/b
  print("Result:",c)
  return([c,op])
```

```python
def check_execution():
  f2=0
  while(f2==0):
    chk = input("Do you want to continue(y or n):")
    if(chk=="n" or chk=="N"):
      flag=1
      f2=1
      print("Program exited")
    elif(chk=="y" or chk=="Y"):
      flag=0
      f2=1
      print("Program continued")
    else:
      print("Wrong selected")
      f2=1
  return(flag)
```

```python
#main program
flag=0
while(flag==0):

  input_lst=input_op() #call the input function to take the inputs and operation

  execution_func(input_lst) #call the execution function to perform operation

  flag=check_execution() #chaeck for the continuation
```
